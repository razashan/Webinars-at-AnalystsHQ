# Time Series Analysis using ARIMA

### What is meant by time series analysis?
Time series analysis is a specific way of analyzing a sequence of data points collected over an interval of time. In time series analysis, analysts record data points at consistent intervals over a set period of time rather than just recording the data points intermittently or randomly.

### What are the four 4 main components of a time series?
Here are the 4 major components:
1. Trend component.
2. Seasonal component.
3. Cyclical component.
4. Irregular component.

### What are the methods of time series?
Times series methods refer to different ways to measure timed data. 
Common types include: <br>
<b>1.</b> Autoregression (AR)<br>
<b>2.</b> Moving Average (MA)<br>
<b>3.</b> Autoregressive Moving Average (ARMA)<br>
<b>4.</b> Autoregressive Integrated Moving Average (ARIMA)<br>
<b>5.</b> Seasonal Autoregressive Integrated Moving-Average (SARIMA).<br>

### What is Trend in Time Series Analysis
In time series analysis, the trend component refers to the long-term pattern or direction of the series. The trend can be linear or nonlinear and can be affected by various factors such as changes in technology, demographics, economic conditions, or other external factors. The trend component can be estimated using statistical methods such as regression analysis or moving averages. Removing the trend component from a time series is often necessary to identify the other components of the series, such as seasonality or cyclical patterns.
<br><br>
In time series analysis, a trend is a long-term change or movement in a time series that persists over a period of time. A trend can be upward (i.e., increasing) or downward (i.e., decreasing).

For example, let's consider a time series of monthly sales data for a company. The data might show that sales have been increasing over the past few years. If this increase in sales is persistent and occurs over a long period of time, then it could be considered a trend.

### What is Seasonal Component in Time Series Analysis
In time series analysis, a seasonal component is a pattern in the data that repeats itself at regular intervals over time. It refers to the regular and periodic fluctuations in the data that occur due to seasonal factors, such as time of year, month, day of the week, or time of day.
<br><br>
For example, a seasonal component can be seen in the sales of winter clothing. During the summer months, the sales of winter clothing are low, while during the winter months, the sales of winter clothing are high. This pattern repeats itself every year, and it is due to the seasonal factor of temperature and weather conditions.
<br><br>
Another example can be seen in the electricity consumption of households. During the summer months, the electricity consumption of households is usually lower, while during the winter months, the electricity consumption is higher. This pattern repeats itself every year, and it is due to the seasonal factor of temperature and weather conditions.
<br><br>
Identifying and modeling the seasonal component in a time series is important in order to better understand the underlying patterns in the data and make more accurate forecasts. It is often modeled using seasonal time series models, such as seasonal ARIMA (SARIMA) or seasonal decomposition of time series (STL).

### What is Cyclical Component in Time Series Analysis 
The cyclical component in time series analysis refers to a type of pattern in the data that occurs over an extended period of time, usually several years or more. These patterns are not as regular as seasonal patterns and do not repeat in a fixed time period. Cyclical patterns are driven by economic, social, or political factors that affect the behavior of the data over time.
<br><br>
For example, in the stock market, there may be a cyclical pattern where the market experiences several years of growth followed by several years of decline. These cycles are driven by economic factors such as interest rates, inflation, and business cycles. Another example could be the housing market where there may be cycles of booms and busts driven by factors such as changes in interest rates, government policies, and population growth.
<br><br>
Identifying and analyzing the cyclical component of a time series can provide valuable insights into the underlying factors driving the data and can help in making informed decisions about future trends and forecasting.

### What is Irregular Component in Time Series Analysis
In time series analysis, the irregular component refers to the random and unpredictable fluctuations or noise present in a time series. It is also known as the random component or residual component. The irregular component is the part of the time series that cannot be explained by the other components, such as trend, seasonality, and cyclical variations.
<br><br>
The irregular component can be modeled as white noise, which is a sequence of uncorrelated random variables with zero mean and constant variance. White noise is a simple and commonly used model for the irregular component of a time series.
<br><br>
For example, in the daily stock prices of a company, the irregular component represents the random fluctuations in the prices that cannot be explained by the trend, seasonality, or cyclical variations. These fluctuations can be due to unexpected news, changes in the market, or other unpredictable factors. By modeling the irregular component as white noise, we can isolate the other components and better understand the underlying patterns in the data.

## What is ARIMA
ARIMA stands for Autoregressive Integrated Moving Average. It is a time series forecasting model that combines <b>autoregression (AR)</b> and <b>moving average (MA)</b> models with differencing to account for non-stationarity. ARIMA models are widely used in analyzing and forecasting time series data, such as economic data, stock prices, and weather patterns.
<br><br>
The <b>"AR"</b> part of ARIMA refers to the autoregressive component, which models the relationship between an observation and a number of lagged observations. The <b>"MA"</b> part of ARIMA refers to the moving average component, which models the relationship between an observation and a residual error from a moving average model applied to lagged observations. The "I" in ARIMA refers to the integrated component, which involves differencing the time series to make it stationary, meaning that its statistical properties such as mean and variance do not change over time.
<br><br>
ARIMA models are specified by three parameters: p, d, and q, where p represents the order of the autoregressive component, d represents the degree of differencing, and q represents the order of the moving average component. The values of these parameters are typically chosen based on the analysis of the autocorrelation and partial autocorrelation functions of the time series data.

### What are the limitations of ARIMA
1. <b>Linearity:</b> ARIMA assumes that the underlying time series is linear, which may not always be the case.

2. <b>Stationarity:</b> ARIMA assumes that the time series is stationary, meaning that its statistical properties do not change over time. However, many real-world time series exhibit non-stationary behavior.

3. <b>Seasonality:</b> ARIMA is not well-suited for modeling time series with complex seasonal patterns, as it can be difficult to determine the appropriate lag order for seasonal terms.

4. <b>Outliers:</b> ARIMA is sensitive to outliers in the data, which can lead to inaccurate forecasts.

5. <b>Computational complexity:</b> ARIMA models can be computationally expensive to fit, especially for long time series or those with high lag orders.

6. <b>Parameter estimation:</b> Estimating the parameters of an ARIMA model can be challenging, as it requires solving a system of complex equations. This can be particularly difficult for models with many parameters or for data sets with missing values.

7. <b>Forecast horizon:</b> ARIMA is generally not well-suited for forecasting more than a few time periods into the future, as the forecasts tend to become increasingly unreliable over longer horizons.

### What is meant by lags in time series analysis 
In time series analysis, lags refer to the time delay between the dependent variable (target variable) and the independent variable (predictor variable) used to predict it. In other words, it is the number of time periods that the predictor variable is shifted forward or backward in time to obtain the best prediction of the target variable.

```python
import pandas as pd
from pandas import datetime as dt
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
from statsmodels.api import tsa

from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
#Loading the data
data = pd.read_csv('/content/bitcoin_USD_close_data-2017-02-09_2018-02-09.csv',
                   parse_dates=['Date'])
print(data.shape)
data.tail()
```

Its clear from the plot that there was a constant upward trend all the way to mid December and then a sharp drop in the price followed by a spark earlier in January, but in general price was going down in 2018.
To better capture the trend of data you can always resample the data by taking biweekly, weekly, monthly or yearly average.

Checking the above graph for monthly average, we can see that it confirms our observation earlier on about the trend of the data. Another way of capturing the trend is through rolling average which we will do below. Basically we take a window of consecutive data points, in our case this is a daily frequency, calculate the average of them, and replace the whole window by that average, either at the extreme right or at the center of the window. So window of 7 is the same as weekly average:

```python
rolling_mean = data.Price.rolling(window=7, center=False).mean() #window of 7 (weekly avg) captures our data better 
plt.figure(figsize=(16,3))
plt.plot(data.Price, color='teal')
plt.plot(rolling_mean, 'red') 
```

## Making the data stationary and autocorrelation

After some initial explorations its now time to create our model. For that we first need to make the series stationary as this is a prerequisite for most of the models. Making the data stationary means taking the trend out of it (de-trending the data) in order to have its statistical properties (i.e. mean, variance) constant over time. This will help the sample to be more predictable by the model since it can be assumed that the statistical properties of the data will be the same in the future as they were in the past. One way of stationarising a time series is through differencing, that is taking the difference of two data points within a specified period, this period is called lag.
In order to find the optimal lag we use autocorerlation method.

Autocorrelation measures the correlation (similarity) between the time series and a lagged version of itself. Below we define varies lag values to see which one causes the highest correlation.

```python
lags = [7, 8, 10, 14, 17, 28, 30, 60, 90, 180] 
autocorrs = [data.Price.autocorr(lag=lag) 
                   for lag in lags]
plt.figure(figsize=(12, 4))
plt.stem(lags, autocorrs)
plt.xlabel("Lag", fontsize=12)
plt.ylabel("Autocorrelation", fontsize=12)
```

Based on the above graph the highest correlation occurs at lag of 7, that means the data repeats a pattern on a weekly basis. This is somewhat true of Bitcoin price as by experience I have seen the price going up throughout the week, normally hitting the highest point on the weekend then coming back down on Monday and repeating more or less the same cycle again

```python
plt.figure(figsize=(16, 6))
plt.plot(data.Price, '-o', color='teal', label="original ts")
plt.plot(data.Price.diff(7), '-o', color='orange', label="differenced ts (lag=7)")
plt.legend(fontsize=12)
```

Now that we know what optimal lag to use to make a relatively stationary version of our original time series we can move on to next step which is training and fitting the models.

```python
prices = data.Price
price_diff = prices.diff(7).dropna() 

plt.figure(figsize=(16,3))
plt.plot(price_diff, "-o", color='orange')
```

Above figure seems stationary enough so optlag=7 chosen for following operations

To start, lets save the stationary data into a new variable and call it time_series:

```python
time_series = price_diff 
```

```python
from statsmodels.tsa.api import AutoReg

ar = AutoReg(time_series, lags=30)
res = ar.fit()
print("Best optimal lag for ar model: {}".format(res.params.shape[0]))
```

<b>AR stands for Auto-Regressive time series model where values are modeled as a linear combination of the p past values therefore it's indicated as AR(p).
p defines the number of past observations that we would want the model to use to predict the next data and is also known as the lag order.</b>

```python

lag = 31
ar = AutoReg(time_series, lags=lag, trend='c')
ar_fit = ar.fit()
ar_prediction = ar_fit.predict(start=lag)
plt.figure(figsize=(16, 4))
plt.plot(time_series, '-o', label='true', color='teal')
plt.plot(ar_prediction, '-o', label='model', color='orange')
plt.legend(fontsize=12);
```

```python
from sklearn.metrics import mean_absolute_error

print('MAE = {0:.3f}'.format(mean_absolute_error(time_series[lag:], ar_prediction)))
print('MAE2 = {0:.3f}'.format(mean_absolute_error(time_series[-90:], ar_prediction[-90:]))) #error only for the last 90 days
```

```python
#7 out of sample prediction with AR 
lag = 47
ar = tsa.AutoReg(time_series,lags=lag)
ar_fit = ar.fit()
ar_forecast = ar_fit.predict(end=len(time_series)+6)[-7:] 
ar_forecast
```

```python
from statsmodels.tsa.arima.model import ARIMA

# 7 out of sample prediction with ARIMA
arima = ARIMA(time_series, order=(7, 0, 7))
arima_fit = arima.fit()
arima_forecast = arima_fit.forecast(steps=7)
arima_forecast
```

```python
#AR model's 7 out of sample predictions
plt.figure(figsize=(16, 4))
plt.plot(time_series, '-o', label="original data", color='teal')
plt.plot(ar_forecast, '--o', label='prediction', color='orange')
plt.legend(fontsize=12)
```
