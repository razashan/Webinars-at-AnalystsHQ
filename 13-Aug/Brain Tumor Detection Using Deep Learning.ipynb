## Introduction to Convolutional Neural Network (CNN)

<img src="Tumor/cnn_full.png" alt="Example Image" width="900">

A Convolutional Neural Network (CNN) is a type of Deep Learning neural network architecture commonly used in Computer Vision. Computer vision is a field of Artificial Intelligence that enables a computer to understand and interpret the image or visual data. 

When it comes to Machine Learning, Artificial Neural Networks perform really well. Neural Networks are used in various datasets like images, audio, and text. Different types of Neural Networks are used for different purposes, for example for predicting the sequence of words we use Recurrent Neural Networks more precisely an LSTM, similarly for image classification we use Convolution Neural networks.

The Convolutional layer applies filters to the input image to extract features, the Pooling layer downsamples the image to reduce computation, and the fully connected layer makes the final prediction. The network learns the optimal filters through backpropagation and gradient descent.

## Common uses for CNNs
There are the following use cases of CNN:<br>
<b>1. Image Classification:</b> CNNs are widely used for image classification tasks, where they can automatically categorize images into different classes or labels. Examples include classifying objects in photographs, identifying diseases in medical images, and recognizing handwritten digits.
<br>
<b>2. Object Detection:</b> CNNs can identify and locate objects within images or video frames. This is particularly useful for tasks like self-driving cars detecting pedestrians, traffic signs, and other vehicles, or security systems detecting intruders.
<br>
<b>3. Image Generation:</b> CNNs can generate new images based on patterns they've learned from a dataset. Variational Autoencoders (VAEs) and Generative Adversarial Networks (GANs) are popular CNN-based architectures for image generation.
<br>
<b>4. Face Recognition:</b> CNNs are used in facial recognition systems to identify and verify individuals based on facial features. These systems are employed in security, authentication, and social media applications.
<br>
<b>5. Image Captioning:</b> CNNs can be combined with Recurrent Neural Networks (RNNs) to generate descriptive captions for images, making it possible for machines to describe images in natural language.
<br>
<b>6. Video Analysis:</b> CNNs can analyze and classify video frames, making them valuable for tasks like action recognition, gesture recognition, and video surveillance.
<br>
<b>7. Medical Imaging:</b> CNNs are used in medical imaging for tasks such as detecting tumors in MRIs, segmenting organs in CT scans, and diagnosing diseases based on medical images.
<br>
<b>8. Natural Language Processing (NLP):</b> While CNNs are primarily used for visual data, they can also be applied to text data. In NLP, CNNs can be used for tasks like text classification, sentiment analysis, and named entity recognition.
<br>
<b>9. Autonomous Vehicles:</b> CNNs play a crucial role in the perception systems of self-driving cars, helping them detect pedestrians, traffic signs, lane markings, and other vehicles on the road.

## Convolution Kernel
<img src = "Tumor/kernel.png">

Convolutional neural networks (CNNs), a kernel, also known as a filter or a convolutional filter, refers to a small matrix used to extract specific features from an input image or volume. The kernel is applied through a process called convolution, which involves sliding the kernel across the input data and performing element-wise multiplication and summation.

Each kernel is designed to capture a specific pattern or feature in the input data. For example, a kernel might be designed to detect edges, corners, or textures. The values within the kernel matrix are typically learned during the training process of the CNN.

## Padding in CNN
<img src="Tumor/9.png" alt="Example Image" width="300">


In convolutional neural networks (CNNs), padding refers to the process of adding extra border pixels around the input image or feature map before applying convolutional operations. Padding is commonly used to control the spatial dimensions of the output feature maps and to address the issue of dimensionality reduction that occurs during convolution.

When a convolution operation is applied without padding, the spatial dimensions of the output feature maps are reduced compared to the input. This reduction occurs because the kernel, or filter, cannot be centered on the pixels at the edges of the input, resulting in incomplete coverage. As a result, the output feature maps have smaller spatial dimension

## Pooling Layer
A pooling layer in a Convolutional Neural Network (CNN) serves to downsample feature maps, reducing their spatial dimensions while preserving essential information. It achieves this by selecting the maximum (max pooling) or average (average pooling) value within small windows as they slide over the feature map. This process enhances the network's ability to detect features regardless of their precise location, leading to better generalization and computational efficiency in subsequent layers.

## Strides
Strides are be used in convolutional kernel and pooling layers. The stride in a pooling layer determines the step size at which the pooling window moves across the input data. Just like in convolutional layers, using larger stride values can lead to downsampled feature maps, effectively reducing the spatial dimensions even further. This can impact the size of the output feature maps and their content, which may influence the network's ability to capture fine details. Using larger stride values in pooling layers can help control the size of the feature maps while potentially reducing computational complexity.

## Flatten Layer
The flatten layer is a crucial component of convolutional neural networks as they connect CNN's to ANN's allowing the neural network models to learn complex patterns and make predictions.
Flatten layer is used to make the multidimensional input one-dimensional, commonly used in the transition from the convolution layer to the full connected layer

## Fully Connected Layer
The Fully connected layer  is used for classifying the input image into a label. This layer connects the information extracted from the previous steps (i.e Convolution layer and Pooling layers) to the output layer and eventually classifies the input into the desired label.


# VGG-19 

## What is VGG-19 Model?
VGG-19, short for Visual Geometry Group 19, is a deep convolutional neural network architecture that was developed by the Visual Geometry Group at the University of Oxford. VGG-19 is known for its simplicity and effectiveness in image classification tasks. It is an extension of the earlier VGG-16 architecture and is named after the number of layers it contains.


1. Architecture Depth: VGG-19 consists of <b>19</b> layers, including <b>16</b> convolutional layers and `<b>3</b> fully connected layers. The "19" in its name refers to the total number of layers.
<br><br>
2. Small Convolutional Kernels: VGG-19 uses relatively small <b>`3x3`</b> convolutional kernels with a stride of <b>1</b> throughout the network. The use of small kernels helps capture local features and patterns in the input images.
<br><br>
3. Max Pooling Layers: VGG-19 uses max pooling layers with a <b>`2x2`</b>window and a stride of <b>2</b> to reduce the spatial dimensions of the feature maps and retain important features while discarding less relevant information.
<br><br>
4. Stacking Convolutional Layers: The architecture is characterized by stacking multiple convolutional layers with small kernels, followed by pooling layers. This deepens the network and allows it to capture more complex and abstract features as the network goes deeper.
<br><br>
5. Fully Connected Layers: After the convolutional and pooling layers, VGG-19 includes three fully connected layers. These layers aggregate the extracted features and make the final classification decision.
<br><br>
6. Activation Function: The network uses the rectified linear unit (ReLU) activation function after each convolutional and fully connected layer. ReLU helps introduce non-linearity to the model.
<br><br>
7. Normalization: VGG-19 does not include batch normalization layers, which are commonly used in modern architectures to accelerate training and improve convergence.
<br><br>
8. ImageNet Pretraining: VGG-19 was initially trained on the ImageNet dataset, a large-scale image classification dataset with millions of labeled images and thousands of classes. The pretrained model can be fine-tuned for specific tasks or used as a feature extractor.
<br><br>
9. Transfer Learning: Due to its effectiveness and pretrained weights, VGG-19 is often used as a base architecture for transfer learning in various computer vision tasks.
<br><br>
While VGG-19 achieved impressive results on image classification tasks, its main drawback is its computational cost and memory requirements due to its depth and the large number of parameters. Despite this, VGG-19 remains an important milestone in the development of deep learning architectures and has contributed to advancing the field of computer vision.

# ResNet-50

## What is ResNet-50 Model?
ResNet stands for Residual Network and is a specific type of convolutional neural network (CNN).
ResNet-50 is a <b>`50`</b>-layer convolutional neural network (<b>`48`</b> convolutional layers, one MaxPool layer, and one average pool layer). Residual neural networks are a type of artificial neural network (ANN) that forms networks by stacking residual blocks.

The main innovation of ResNet-50 lies in its use of residual blocks, also known as skip connections or shortcut connections. These blocks introduce a concept called "residual learning," which allows very deep networks to be trained more effectively. 

1. Residual Blocks: A residual block consists of multiple convolutional layers followed by shortcut connections. These shortcut connections skip one or more layers and directly add the input from a previous layer to the output of the residual block. This allows the network to learn the residual (difference) between the input and the output, making it easier for the network to learn identity mappings and preventing the vanishing gradient problem. The skip connections enable the training of much deeper networks without suffering from degradation in performance.
<br><br>
2. Identity Mapping: In traditional neural networks, deeper layers can sometimes result in decreased accuracy due to the vanishing gradient problem. ResNet-50's residual blocks enable the network to learn the identity mapping, meaning that if the optimal transformation is close to an identity mapping, the residual block can learn to make small adjustments, and the network can still learn the desired transformation.
<br><br>
3. Architecture: ResNet-50 consists of 50 layers, including convolutional layers, residual blocks, and fully connected layers. It begins with a convolutional layer followed by max-pooling, and then a series of residual blocks of varying depths. The architecture also employs global average pooling and a fully connected layer for classification.
<br><br>
4. Pre-trained Models: ResNet-50, along with other ResNet variants, has been pre-trained on large image datasets like ImageNet. This pre-training allows the network to learn a rich set of feature representations that can be fine-tuned or used as feature extractors for various image-related tasks, such as image classification, object detection, and more.

## Data Augmentation
<img src="Tumor/Original Image.jpg" width=600>
Data augmentation is a technique commonly used in machine learning and deep learning to artificially increase the diversity and quantity of training data by applying various transformations to the existing data. The goal of data augmentation is to improve the performance and generalization ability of a machine learning model by exposing it to a wider range of scenarios, variations, and conditions that it might encounter in real-world situations.
<br>
<br>
Data augmentation helps to prevent overfitting, where a model performs well on the training data but fails to generalize to unseen data. By exposing the model to augmented data during training, it becomes more robust and adaptable to variations in the input data. This is particularly important when the available training data is limited or when the model's performance can benefit from exposure to a wider range of scenarios.

```python
start = timeit.default_timer()
path = "/kaggle/input/brain-mri-images-for-brain-tumor-detection/brain_tumor_dataset"

image_generator = ImageDataGenerator(
    rescale=1./255, 
    horizontal_flip=2, 
    vertical_flip=2, 
    validation_split=0.2)   

training_set = image_generator.flow_from_directory(batch_size=16,
                                                 directory=path,
                                                 shuffle=True,
                                                 target_size=(224,224), 
                                                 subset="training",
                                                 color_mode='rgb',
                                                 class_mode='binary')

testing_set = image_generator.flow_from_directory(batch_size=16,
                                                 directory=path,
                                                 shuffle=True,
                                                 target_size=(224,224), 
                                                 subset="validation",
                                                 color_mode='rgb',
                                                 class_mode='binary')
```

```python
yes_path = path+"/yes/"+os.listdir(path+"/yes/")[4]
tumour = imread(yes_path)
plt.imshow(tumour)
```

```python
no_path = path+"/no/"+os.listdir(path+"/no/")[4]
tumour = imread(yes_path)
plt.imshow(tumour)
```

```python
testing_set.class_indices
```

## Training VGG Model

```python
model = VGG19(
      input_shape = (224,224,3),
      include_top = False,
      weights = 'imagenet'
    )
```

input_shape = (224,224,3) indicates that the VGG19 model expects input images with a resolution of 224x224 pixels and three color channels (RGB). This information is necessary for the model to correctly define the dimensions of the input layer and process the data appropriately during training and inference.

```python
for layers in model.layers:
    layers.trainable = False
```

## What is Dropout Layer?
Dropout is a regularization technique commonly used in neural networks to prevent overfitting. Overfitting occurs when a neural network learns to perform exceptionally well on the training data but fails to generalize to new, unseen data. Dropout helps to address this issue by randomly "dropping out" (deactivating) a fraction of neurons during each training iteration.

## What is Activation Function?
An activation function is a crucial component of a neural network or any artificial neuron-like model. It defines the output of a neuron or a layer of neurons based on the weighted sum of its inputs. The activation function introduces non-linearity to the model, allowing it to learn and represent complex relationships in the data.

## What is Binary Cross-Entropy?
Binary Cross-Entropy, also known as Binary Logarithmic Loss or Binary Cross-Entropy Loss, is a commonly used loss function in machine learning, particularly in binary classification tasks. It measures the dissimilarity between the predicted probability distribution and the true binary labels of the data. This loss function is often used in scenarios where each input belongs to one of two classes (e.g., positive/negative, 1/0).

## What is Optimizer?
An optimizer is a crucial component of the training process in machine learning, especially in the context of training neural networks. It is an algorithm or method that adjusts the parameters of a model, typically the weights and biases of the model's neurons, in order to minimize a predefined loss function. The goal of optimization is to find the set of parameters that result in the best possible performance of the model on the training data and, ideally, on unseen data as well.

```python
x = Flatten()(model.output)
x = Dropout(0.4)(x)
x = Dense(1, activation = "sigmoid")(x)
model = keras.Model(model.input, x)
model.compile(loss = "binary_crossentropy", optimizer = "adam", metrics = "accuracy")
model.summary()
```

```python
hist = model.fit_generator(training_set,validation_data = testing_set, epochs = 60)
```

```python
hist = hist.history
```

```python
plt.plot(hist["accuracy"])
plt.plot(hist["val_accuracy"])
plt.title("VGG-19 Plot")
plt.legend(["train","test"])
plt.xlabel("epoch")
plt.ylabel("accuracy")
```

```python
plt.plot(hist["loss"])
plt.plot(hist["val_loss"])
plt.title("VGG-19 loss")
plt.legend(["train","test"])
plt.xlabel("epoch")
plt.ylabel("loss")
```

```python
# Access the training and validation accuracy values
train_accuracy = hist["accuracy"]
val_accuracy = hist["val_accuracy"]

# Get the overall accuracy (final accuracy after training)
overall_train_accuracy = train_accuracy[-1]
overall_val_accuracy = val_accuracy[-1]

# Print the overall accuracy
print(f"Overall Training Accuracy: {overall_train_accuracy:.4f}")
print(f"Overall Validation Accuracy: {overall_val_accuracy:.4f}")

evaluation_result = model.evaluate(testing_set)

# Extract the accuracy from the evaluation result
overall_accuracy = evaluation_result[1]

# Print the overall accuracy
print(f"Overall Model Accuracy: {overall_accuracy:.4f}")
```

```python
import cv2
import numpy as np

def preprocess_input_image(image_path):
    img = cv2.imread(image_path)
    img = cv2.resize(img, (224, 224))  # Resize to match the model's input size
    img = img / 255.0  # Normalize the pixel values
    img = np.expand_dims(img, axis=0)  # Add batch dimension
    return img
```
